```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
BA = "Resources/BA.csv"
UAL = "Resources/UAL.csv"
DAL = "Resources/DAL.csv"
AAL = "Resources/AAL.csv"
Covid = "Resources/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv"
```

```python
BA_df = pd.read_csv(BA)
UAL_df = pd.read_csv(UAL)
DAL_df = pd.read_csv(DAL)
AAL_df = pd.read_csv(AAL)
Covid_df = pd.read_csv(Covid)
```

```python
airlines_df1 = pd.DataFrame(BA_df, columns=[
                        "Date", "Open", "High", "Low", "Close"])
airlines_df2 = pd.DataFrame(UAL_df, columns=[
                        "Date", "Open", "High", "Low", "Close"])
airlines_df3 = pd.DataFrame(DAL_df, columns=[
                        "Date", "Open", "High", "Low", "Close"])
airlines_df4 = pd.DataFrame(AAL_df, columns=[
                        "Date", "Open", "High", "Low", "Close"])
Covid_df2 = pd.DataFrame(Covid_df, columns=[
                        "Date", "state", "tot_cases"])
```

```python
merged_df = pd.merge(airlines_df1, airlines_df2, how='outer')
merged_df
```

```python
merged_df2 = pd.merge(airlines_df3, airlines_df4, how='outer')
merged_df2
```

```python
merged_df3 = pd.merge(merged_df, airlines_df2)
merged_df3
```

```python
final_merged = pd.concat([merged_df3, Covid_df2], axis=1)
final_merged
```
